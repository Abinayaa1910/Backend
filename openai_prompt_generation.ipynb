```python
def generate_prompt(HDBSCAN_CatNum_Label_1):
    prompts = {
        0: f"Craft a warm, community-focused promo message for a mid-term male customer living in Central or Choa Chu Kang. ",
        1: f"Write a dynamic intro offer message for a new male customer who just joined in late 2024, from Geylang or Queenstown. ",
        2: f"Create a loyalty reward message for a long-time customer from Bukit Panjang or Queenstown. ",
        3: f"Write an elegant, premium-style promo for a mid-term female from Bukit Timah or Geylang. ",
        4: f"Draft an engaging welcome promo for a new female customer from Bukit Batok or Serangoon, who joined in 2025. ",
        5: f"Generate an upselling promo for a mid-term female from Tampines or Queenstown. Use this note: ",
        6: f"Create a luxury package message for a loyal customer from Sembawang or Bukit Merah. Use the note as context: ",
        7: f"Write a thank-you and referral promo for a loyal female customer from Bedok or Tampines. Reference: ",
        8: f"Create a family-friendly offer for a mid-term male customer from Hougang or Jurong East. Note: ",
    }
    return prompts.get(HDBSCAN_CatNum_Label_1, f"Write a cruise promotion message based on this note: ")
```

```python
import pandas as pd

# Load the exported DataFrame
df = pd.read_csv("clustered_customers.csv")

# Preview to make sure it loaded correctly
df.head()
```

```python
import time
```

```python
generated_outputs = []

for index, row in df.iterrows():
    prompt = generate_prompt(row['HDBSCAN_CatNum_Label_1'])

    try:
        response = client.chat.completions.create(
            model="gpt-3.5-turbo",  # change to "gpt-4" for testing later
            messages=[
                {"role": "system", "content": "You are a marketing assistant for a cruise booking company."},
                {"role": "user", "content": prompt}
            ],
            temperature=0.7,
            max_tokens=100,
            top_p=0.9
        )
        content = response.choices[0].message.content
        generated_outputs.append(content)

        # OPTIONAL: print output
        print(f"Customer {index} â†’ Cluster {row['HDBSCAN_CatNum_Label_1']}:\n{content}\n")

        time.sleep(1)  # Avoid hitting rate limits

    except Exception as e:
        print(f"Error at index {index}: {e}")
        generated_outputs.append("ERROR")
```
